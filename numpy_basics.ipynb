```python
#4
player = np.array([45,72,33,61,88,54,40,76,92,58])
print(player)
```

```python
mean_player = np.mean(player)
print(mean_player)
```

```python
std_player = np.std(player)
print(std_player)
```

```python
var_player = np.var(player)
print(var_player)
```

```python
cv = (std_player/mean_player)*100
print(cv)
```

```python
if cv > 75:
  print('Player is consistent')
else:
  print('Player is inconsistent')
```

```python
#5
runs = np.array([1,4,0,6,2,1,4,0,1,6,2,1])
print(runs)
```

```python
total_runs = np.sum(runs)
print(total_runs)
```

```python
boundary_runs = np.sum(runs[(runs == 4) | (runs == 6)])
print(boundary_runs)
```

```python
boundary_percent = (boundary_runs / total_runs) * 100
print(boundary_percent)
```
